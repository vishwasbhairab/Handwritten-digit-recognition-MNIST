```python
import cv2
import numpy as np
import tensorflow as tf
```

```python
# Load the trained CNN model
model = tf.keras.models.load_model("mnist_cnn.keras")
```

```python
# Initialize webcam
cap = cv2.VideoCapture(0)

# Define the ROI (Region of Interest) size
ROI_SIZE = 300

while True:
    ret, frame = cap.read()
    if not ret:
        break

    # Flip the frame horizontally (so it acts like a mirror)
    frame = cv2.flip(frame, 1)

    # Get frame dimensions
    height, width, _ = frame.shape

    # Define a square ROI for drawing digits
    x1, y1 = width // 2 - ROI_SIZE // 2, height // 2 - ROI_SIZE // 2
    x2, y2 = width // 2 + ROI_SIZE // 2, height // 2 + ROI_SIZE // 2

    # Extract the ROI
    roi = frame[y1:y2, x1:x2]

    # Convert ROI to grayscale
    gray = cv2.cvtColor(roi, cv2.COLOR_BGR2GRAY)

    # Apply thresholding to create a binary image
    _, thresh = cv2.threshold(gray, 100, 255, cv2.THRESH_BINARY_INV)

    # Resize to (28x28) to match the MNIST input size
    digit = cv2.resize(thresh, (28, 28))

    # Normalize the image (convert to float and scale between 0-1)
    digit = digit.astype("float32") / 255.0

    # Reshape to match CNN input format (batch_size, height, width, channels)
    digit = digit.reshape(1, 28, 28, 1)

    # Predict the digit
    prediction = model.predict(digit)
    predicted_digit = np.argmax(prediction)
    confidence = np.max(prediction) * 100

    # Draw the ROI rectangle on the frame
    cv2.rectangle(frame, (x1, y1), (x2, y2), (0, 255, 0), 2)

    # Display prediction results
    text = f"Digit: {predicted_digit} ({confidence:.2f}% confidence)"
    cv2.putText(frame, text, (x1, y1 - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.8, (0, 255, 0), 2)

    # Show the processed frame
    cv2.imshow("Digit Recognition", frame)
    cv2.imshow("Thresholded Digit", thresh)

    # Exit loop when 'q' is pressed
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

# Release resources
cap.release()
cv2.destroyAllWindows()
```
